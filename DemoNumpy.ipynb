```python
import numpy as np
```

```python
a1=np.array([[2,3,4],
             [1,4,7]])
a2=np.array([[1,1,1],[1,1,1]])
add=np.add(a1,a2)
print(add)


a=np.array([1,2,3,4])
b=np.array([9,8,7,6])
print("a:",a)
print("b:",b)
add1=np.add(a,b)
print("a+b:",add1)

sub=np.subtract(b,a)
print("a-b:",sub)

mul=np.multiply(a,b)
print("a-b:",mul)

div=np.divide(a,b)
print("a/b:",div)

mod=np.mod(a,b)
print("a%b:",mod)
```
