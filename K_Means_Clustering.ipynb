```python
'''
K-Means Clustering is an unsupervised machine learning algorithm used to group
similar data points into clusters based on their features.

It is called unsupervised because the dataset does not contain labeled output values.
The algorithm automatically discovers patterns in the data.

K → Number of clusters (groups) we want to create

Means → The average (centroid) of data points in each cluster

| Cluster   | Meaning                    |
| --------- | -------------------------- |
| Cluster 0 | High income, high spending |
| Cluster 1 | High income, low spending  |
| Cluster 2 | Low income, high spending  |
| Cluster 3 | Average customers          |
| Cluster 4 | Low income, low spending   |

Data Points

● ● ●        ○ ○ ○        ▲ ▲ ▲

Cluster 1    Cluster 2    Cluster 3

Why we use this?

It helps discover hidden patterns in unlabeled data.

Commonly used for customer segmentation, market analysis, and pattern detection.

The algorithm works by assigning data points to the nearest cluster centroid and updating the centroid iteratively.

'''
```

```python
# ---------------------------------------------------------
# 1. IMPORT REQUIRED LIBRARIES
# ---------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# ---------------------------------------------------------
# 2. LOAD DATASET
# ---------------------------------------------------------

data = pd.read_csv("/content/sample_data/customer_segmentation_dataset.csv")

print("Dataset Preview:")
print(data.head())


# ---------------------------------------------------------
# 3. SELECT FEATURES FOR CLUSTERING
# ---------------------------------------------------------

# Using Annual Income and Spending Score for clustering
X = data[['Annual_Income','Spending_Score']]


# ---------------------------------------------------------
# 4. APPLY K-MEANS CLUSTERING
# ---------------------------------------------------------

# Create model with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
# random_state = 42 K-Means starts by selecting random centroids.
# Train the model
kmeans.fit(X)

# Assign cluster labels
data['Cluster'] = kmeans.predict(X)

print("\nClustered Data:")
print(data.head(50))


# ---------------------------------------------------------
# 5. VISUALIZE CLUSTERS
# ---------------------------------------------------------

plt.figure(figsize=(8,6))

plt.scatter(data['Annual_Income'], data['Spending_Score'],
            c=data['Cluster'], cmap='rainbow')

plt.xlabel("Annual Income ($)")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation using K-Means Clustering")

plt.show()


# ---------------------------------------------------------
# 5. VISUALIZE CLUSTERS
# ---------------------------------------------------------
'''
o → circle

s → square

^ → triangle

D → diamond

* → star

'''

plt.figure(figsize=(8,6))

plt.scatter(data[data['Cluster']==0]['Annual_Income'],
            data[data['Cluster']==0]['Spending_Score'],
            marker='o', label='Cluster 0')

plt.scatter(data[data['Cluster']==1]['Annual_Income'],
            data[data['Cluster']==1]['Spending_Score'],
            marker='s', label='Cluster 1')

plt.scatter(data[data['Cluster']==2]['Annual_Income'],
            data[data['Cluster']==2]['Spending_Score'],
            marker='^', label='Cluster 2')

plt.scatter(data[data['Cluster']==3]['Annual_Income'],
            data[data['Cluster']==3]['Spending_Score'],
            marker='D', label='Cluster 3')

plt.scatter(data[data['Cluster']==4]['Annual_Income'],
            data[data['Cluster']==4]['Spending_Score'],
            marker='*', label='Cluster 4')

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation using K-Means Clustering")


plt.legend()

plt.show()
```
